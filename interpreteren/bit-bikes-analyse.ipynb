Import dependencies die nodig zijn om de visualisatie te maken.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Laad de data in

```python
data_2020 = pd.read_csv('data/202007-divvy-tripdata.csv')
data_2021 = pd.read_csv('data/202107-divvy-tripdata.csv')
data_2022 = pd.read_csv('data/202207-divvy-tripdata.csv')

data_2020['started_at'] = pd.to_datetime(data_2020['started_at'], format="%d/%m/%Y %H:%M")
data_2020['ended_at'] = pd.to_datetime(data_2020['ended_at'], format="%d/%m/%Y %H:%M")

data_2021['started_at'] = pd.to_datetime(data_2021['started_at'], format="%Y-%m-%d %H:%M:%S")
data_2021['ended_at'] = pd.to_datetime(data_2021['ended_at'], format="%Y-%m-%d %H:%M:%S")

data_2022['started_at'] = pd.to_datetime(data_2022['started_at'], format="%Y-%m-%d %H:%M:%S")
data_2022['ended_at'] = pd.to_datetime(data_2022['ended_at'], format="%Y-%m-%d %H:%M:%S")

data_2020_2021_2022 = pd.concat([data_2020, data_2021, data_2022])
```

# Ritten
We gaan de data van 2020 bekijken doormiddel van een lijndiagram. Ik kies hiervoor zodat we goed kunnen zien hoevel ritten er per dag zijn gemaakt in 2020.

```python
rides_per_day = data_2020.groupby(data_2020['started_at'].dt.date).size()

sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=rides_per_day, marker='o', linewidth=2)
plt.xlabel('Datum')
plt.ylabel('Aantal ritten')
plt.title('Aantal ritten per dag 2020')
plt.xticks(rotation=45)
plt.show()
```

We gaan de data van 2021 bekijken doormiddel van een lijndiagram. Ik kies hiervoor zodat we goed kunnen zien hoevel ritten er per dag zijn gemaakt in 2021.


```python
rides_per_day = data_2021.groupby(data_2021['started_at'].dt.date).size()

sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=rides_per_day, marker='o', linewidth=2)
plt.xlabel('Datum')
plt.ylabel('Aantal ritten')
plt.title('Aantal ritten per dag 2021')
plt.xticks(rotation=45)
plt.show()
```

We gaan de data van 2022 bekijken doormiddel van een lijndiagram. Ik kies hiervoor zodat we goed kunnen zien hoevel ritten er per dag zijn gemaakt. in 2022.


```python
rides_per_day = data_2022.groupby(data_2022['started_at'].dt.date).size()

sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=rides_per_day, marker='o', linewidth=2)
plt.xlabel('Datum')
plt.ylabel('Aantal ritten')
plt.title('Aantal ritten per dag 2022')
plt.xticks(rotation=45)
plt.show()
```

We hebben zojuist hierboven een aantal lijndiagram gezien die tonen per jaar de ritten tonen. Hieronder zien we een lijndiagram van de afgelopen 3 jaar naast elkaar gezet zodat we een beter beeld krijgen van de verschillen tussen de jaren.

```python
rides_per_day_2020 = data_2020.groupby(data_2020['started_at'].dt.strftime('%m-%d')).size()
rides_per_day_2021 = data_2021.groupby(data_2021['started_at'].dt.strftime('%m-%d')).size()
rides_per_day_2022 = data_2022.groupby(data_2022['started_at'].dt.strftime('%m-%d')).size()

sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=rides_per_day_2020, marker='o', linewidth=2, label='2020')
sns.lineplot(data=rides_per_day_2021, marker='o', linewidth=2, label='2021')
sns.lineplot(data=rides_per_day_2022, marker='o', linewidth=2, label='2022')
plt.xlabel('Datum')
plt.ylabel('Aantal ritten')
plt.title('Aantal ritten per dag (2020-2022)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

We zien hierboven een duidelijk verschil tussen 2020 en 2021 toen was de hoeveelheid ritten bijna de helft op zijn peak maar we zien in 2022 bijna de zelfde hoeveelheid ritten als in 2021. Dit is een goed teken voor Bit Bikes.

# Conclusie trips
Aan de hoeveelheid ritten te zien is he hoeveelheid ongeveer gelijk in 2021 tegenover 2022 zonder verwachte groei door overnamens of andere factoren. Is mijn advies om geen fietsen te verkopen maar eerder mochten er fietsen kapot gaan of gestolen worden om deze te vervangen en de hoeveelheid fietsen op het zelfde aantal te houden.

```python
# Membership
```

```python
casual_per_year = [data_2020[data_2020['member_casual'] == 'casual']['member_casual'].count(),
                   data_2021[data_2021['member_casual'] == 'casual']['member_casual'].count(),
                   data_2022[data_2022['member_casual'] == 'casual']['member_casual'].count()]

member_per_year = [data_2020[data_2020['member_casual'] == 'member']['member_casual'].count(),
                   data_2021[data_2021['member_casual'] == 'member']['member_casual'].count(),
                   data_2022[data_2022['member_casual'] == 'member']['member_casual'].count()]

years = [2020, 2021, 2022]

plt.figure(figsize=(10, 6))
plt.plot(years, casual_per_year, marker='o', linewidth=2, label='Casual')
plt.plot(years, member_per_year, marker='o', linewidth=2, label='Member')

plt.xlabel('Jaar')
plt.ylabel('Aantal')
plt.title('Aantal casual en member per jaar')
plt.xticks(years)
plt.legend()

plt.show()

bar_width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(years, casual_per_year, width=bar_width, label='Casual')
plt.bar(years, member_per_year, width=bar_width, label='Member', bottom=casual_per_year)

plt.xlabel('Jaar')
plt.ylabel('Aantal')
plt.title('Aantal casual en member per jaar')
plt.xticks(years)
plt.legend()

plt.show()
```

# Conclusie membership
We zien een groei in aantal members wat we zien dat meerdere casuals overgegaan zijn naar een membership wat een goed teken is voor Bit Bikes. We zien hierdoor dat de omzet van Bit Bikes ook zal groeien. Er zijn wat minder casuals in 2022 maar dit is niet een groot verschil en zal niet veel invloed hebben op de omzet van Bit Bikes. Omdat er meer members zijn in 2022 zal er meer omzet zijn dan in 2021. We zien de hoeveelheid gebruikers ongeveer gelijk gebleven is in 2021 en 2022 maar dat er meer members zijn in 2022. Mijn advies is om de membership te promoten en de casuals te overtuigen om een membership te nemen.

# FietsTypes

```python
docked_bike_per_year = [data_2020[data_2020['rideable_type'] == 'docked_bike']['rideable_type'].count(),
                   data_2021[data_2021['rideable_type'] == 'docked_bike']['rideable_type'].count(),
                   data_2022[data_2022['rideable_type'] == 'docked_bike']['rideable_type'].count()]

classic_bike_per_year = [data_2020[data_2020['rideable_type'] == 'classic_bike']['rideable_type'].count(),
                   data_2021[data_2021['rideable_type'] == 'classic_bike']['rideable_type'].count(),
                   data_2022[data_2022['rideable_type'] == 'classic_bike']['rideable_type'].count()]
electric_bike_per_year = [data_2020[data_2020['rideable_type'] == 'electric_bike']['rideable_type'].count(),
                   data_2021[data_2021['rideable_type'] == 'electric_bike']['rideable_type'].count(),
                   data_2022[data_2022['rideable_type'] == 'electric_bike']['rideable_type'].count()]

years = [2020, 2021, 2022]

plt.figure(figsize=(10, 6))
plt.plot(years, docked_bike_per_year, marker='o', linewidth=2, label='Docked')
plt.plot(years, classic_bike_per_year, marker='o', linewidth=2, label='Classic')
plt.plot(years, electric_bike_per_year, marker='o', linewidth=2, label='Electric')

plt.xlabel('Jaar')
plt.ylabel('Aantal')
plt.title('Aantal docked, electric en classic fietsen per jaar')
plt.xticks(years)
plt.legend()

plt.show()
```

# Conclusie fietstypes
We zien een duidelijke groei van Electric bikes en een daling van docked bikes. Dit is een goed teken voor Bit Bikes omdat de electric bikes meer omzet genereren dan de docked bikes. We zien ook een kleine groei in classic bikes maar dit is niet een groot verschil. Mijn advies is om meer electric bikes te kopen en de docked bikes te verkopen. Eventueel kunnen de classic bikes ook verkocht worden maar dit is niet nodig.
