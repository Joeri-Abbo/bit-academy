# Manipuleren voor Analisten → 6.Data verrijken → 2.Nog meer bejaarde bulldozers

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bulldozers.csv')
```

Cast de kolom 'saledate' naar het datatype datetime

```python
df['saledate'] = pd.to_datetime(df['saledate'])
```

Haal het jaar uit de 'saledate' en voeg deze toe als extra kolom 'SalesYear'

```python
df['SalesYear'] = df['saledate'].dt.year
```

Bereken de leeftijd in jaren van de bulldozers

```python
df['Age'] = df['SalesYear'] - df['YearMade']
df = df[df['Age'] <= 100]
```

Maak een scatterplot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['SalePrice'], alpha=0.5)
plt.title('Leeftijd van Bulldozers vs. Verkoopprijs')
plt.xlabel('Leeftijd (jaren)')
plt.ylabel('Verkoopprijs ($)')
plt.grid(True)
plt.show()
```
