# Manipuleren voor Analisten → 5.Buitenbeentjes → 2.Outliers spotten!

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bulldozers.csv')
```

Eerste plot - Het aantal bulldozers dat in een periode gemaakt is.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['YearMade'], bins=50, edgecolor='k')
plt.title('Het aantal bulldozers dat in een periode gemaakt is.')
plt.xlabel('Jaar van productie')
plt.ylabel('Aantal')
plt.show()
```

Verwijderen van outliers (jaartallen onder 1960)

```python
df = df[df['YearMade'] >= 1960]
df = df[df['YearMade'] <= 2020]
```

Tweede plot - Het aantal bulldozers dat in een periode gemaakt is.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['YearMade'], bins=20, edgecolor='k')
plt.xticks(range(1920, 2020, 10)) 
plt.title('Het aantal bulldozers dat in een periode gemaakt is.')
plt.xlabel('Jaar')
plt.ylabel('Aantal bulldozers')
plt.show()
```
