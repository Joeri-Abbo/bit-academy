# Gezonde maaltijd
we gaan een barchart maken voor de maaltijden met het mist aantal calorieën per categorie

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Laad het CSV-bestand in een pandas DataFrame

```python
df = pd.read_csv('mc_menu_final.csv')
```

Vind per maaltijdcategorie de maaltijd met het minste aantal calorieën

```python
min_calorie_meals = df.groupby('Category')['Calories'].idxmin()  
min_calorie_df = df.loc[min_calorie_meals, ['Category', 'Item', 'Calories']]  
```

Maak een barchart met aantal calorieën voor de maaltijden met het minste aantal calorieën per categorie

```python
plt.figure(figsize=(12, 6))
plt.bar(min_calorie_df['Category'], min_calorie_df['Calories'])
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Calories')
plt.legend(['Calories'])

plt.show()
```
