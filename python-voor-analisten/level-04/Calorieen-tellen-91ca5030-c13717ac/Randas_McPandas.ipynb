# Calorieen tellen
We gaa Calorieen tellen van verschillende de categorie "Chicken & Fish" producten

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Laad het CSV-bestand in een pandas DataFrame

```python
df = pd.read_csv('mc_menu_final.csv')
```

Maak een barchart met Item op de x-as en Calories op de y-as voor de categorie "Chicken & Fish"

```python
chicken_fish_df = df[df['Category'] == 'Chicken & Fish']
plt.figure(figsize=(12, 6))
plt.bar(chicken_fish_df['Item'], chicken_fish_df['Calories'])
plt.xticks(rotation=90)
plt.xlabel('Item')
plt.legend(['Calories'])
plt.ylabel('Calories')
plt.show()
```

Verwijder rijen waar het aantal calorien NaN is

```python
chicken_fish_df = chicken_fish_df.dropna(subset=['Calories'])
```

Plot een barchart met de eigenschappen van de barchart uit stap 1 met de bijgewerkte DataFrame

```python
plt.figure(figsize=(12, 6))
plt.bar(chicken_fish_df['Item'], chicken_fish_df['Calories'])
plt.xticks(rotation=90)
plt.xlabel('Item')
plt.legend(['Calories'])

plt.ylabel('Calories')
plt.show()
```
