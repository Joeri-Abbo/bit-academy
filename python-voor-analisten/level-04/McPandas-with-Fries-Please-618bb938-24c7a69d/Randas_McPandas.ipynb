# McPandas with Fries, Please
We gaan barchart maken met over Smoothies & Shakes en hoevel fat deze bevatten

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Laad het csv bestand in

```python
df = pd.read_csv('mc_menu_final.csv')
```

Maak een barchart met dezelfde voorwaarden, alleen gebruik nu enkel producten van de categorie "Smoothies & Shakes"

```python
smoothies_df = df[df['Category'] == 'Smoothies & Shakes']
plt.figure(figsize=(12, 6))
plt.bar(smoothies_df['Item'], smoothies_df['Total Fat'])
plt.xticks(rotation=90)
plt.xlabel('Item')
plt.ylabel('Total Fat')
plt.title('Total Fat in McPandas Smoothies & Shakes')
plt.legend(['Total Fat'])
plt.show()
```

Identificeer items met mogelijke foutieve waarden

```python
threshold = 30  # de drempelwaarde voor vetgehalte
potential_errors = smoothies_df[smoothies_df['Total Fat'] > threshold]['Item'].tolist()
```

Toon de gedentificeerde items met mogelijke foutieve waarden

```python
print("Mogelijke foutieve waarden in het vetgehalte:")
for item in potential_errors:
    print("- " + item)
```
