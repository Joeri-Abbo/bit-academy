In de vorige exercise heb je kennis gemaakt met matplotlib. Dit is een erg uitgebreide module waar je tot op pixelniveau je plots kan aanpassen. Dat biedt dus heel veel vrijheid, maar kan ook zorgen voor ingewikkelde en onnodig lange code. Daarom is het verstandig om, als je data toch al in pandas staat, een module zoals seaborn te gebruiken waar mogelijk.

Je kan matplotlib wel goed gebruiken om meerdere plots in één figuur te zetten. In deze exercise gaan we 3 plots maken met seaborn, om die vervolgens met matplotlib in hetzelfde figuur te plaatsen. De plots die samen in één figuur staan noemen we axes of subplots van dat figuur.

Import dependencies die nodig zijn om de visualisatie te maken.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Laad de data in

```python
data = pd.read_csv('House_Rent_Dataset.csv')
```

Maak de visualisatie

```python
import numpy as np
color_palette = plt.cm.get_cmap('tab20', len(data['City'].unique()))

state_1_df = data[data['State'] == 1]
state_2_df = data[data['State'] == 2]
state_3_df = data[data['State'] == 3]

state_1_city_counts = state_1_df['City'].value_counts()
state_2_city_counts = state_2_df['City'].value_counts()
state_3_city_counts = state_3_df['City'].value_counts()

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

state_1_ax = axes[0].bar(state_1_city_counts.index, state_1_city_counts.values, color=color_palette(np.arange(len(state_1_city_counts))))
axes[0].set_ylabel('The number of worn down houses in each city')
axes[0].set_xlabel('Rent (INR)')
axes[0].set_title('Staat 1')

state_2_ax = axes[1].bar(state_2_city_counts.index, state_2_city_counts.values, color=color_palette(np.arange(len(state_2_city_counts))))
axes[1].set_ylabel('Number of houses')
axes[1].set_xlabel('Rent (INR)')
axes[1].set_title('The number of averagly maintained houses in each city')

state_3_ax = axes[2].bar(state_3_city_counts.index, state_3_city_counts.values, color=color_palette(np.arange(len(state_3_city_counts))))
axes[2].set_ylabel('Number of new or as new houses in each city')
axes[2].set_xlabel('Rent (INR)')
axes[2].set_title('As new')

fig.tight_layout()
plt.show()
```
