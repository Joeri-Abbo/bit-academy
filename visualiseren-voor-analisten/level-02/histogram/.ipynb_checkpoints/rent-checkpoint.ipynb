# Histogram
We gaan voor rent meerdere histogram maken zie hieronder de stappen om dit te doen.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Laad het CSV-bestand in een DataFrame

```python
df = pd.read_csv("House_Rent_Dataset.csv")
```

Plot de verdeling van de kolom "Rent" met standaard aantal bins

```python
sns.histplot(data=df, x="Rent")
plt.title("Verdeling van de huurprijzen")
plt.xlabel("Huurprijs")
plt.ylabel("Aantal")
plt.show()
```

Plot de verdeling van de kolom "Rent" met 50 bins

```python
sns.histplot(data=df, x="Rent", bins=50)
plt.title("Verdeling van de huurprijzen (50 bins)")
plt.xlabel("Huurprijs")
plt.ylabel("Aantal")
plt.show()
```
