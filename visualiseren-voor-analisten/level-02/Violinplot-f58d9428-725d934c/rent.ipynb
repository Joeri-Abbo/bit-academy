Import dependencies die nodig zijn om de visualisatie te maken.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Laad de data in

```python
df = pd.read_csv('House_Rent_Dataset.csv')
```

Filter huizen in staat 1 en 2

```python
df_filtered = df[df['State'].isin([1, 2])]
```

Maak de  violinplot

```python
plt.figure(figsize=(10,6))
# Map label 1 to 'Condition 1', label 2 to 'Condition 2'
df_filtered['State'] = df_filtered['State'].map({1: 'Worn down state', 2: 'Average state '})
plt.legend(loc='upper right')
sns.violinplot(x='City', y='Rent', hue='State', data=df_filtered, split=True)

plt.show()
```

# Alle huizen hebben een hogere huurprijs bij een betere staat, behalve in de stad Hyderabad.
Dit is hieboven duidelijk te zien. De huizen in Hyderabad hebben een hogere huurprijs bij een slechtere staat.

```python


# Add swarmplot
plt.figure(figsize=(10,6))
sns.violinplot(x='City', y='Rent', hue='State', data=df_filtered, split=True)
sns.swarmplot(x='City', y='Rent', hue='State', data=df_filtered, color='black')
plt.legend().remove()

plt.show()
```

 # Probeer op basis van de laatste grafiek te verklaren waarom Hyderabad niet dezelfde trend vertoont als de andere steden.

De hoeveelheid huizen in Hyderabad is veel minder dan in de andere steden. Hierdoor is de kans dat er een uitschieter in de data zit groter. Dit is ook te zien in de grafiek. De huizen in Hyderabad hebben een hogere huurprijs bij een slechtere staat. Dit is een uitschieter in de data. De huizen in de andere steden hebben een hogere huurprijs bij een betere staat. Dit is de trend die we verwachten.
