Import the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Laad in de dataset en genereer een lineplot

```python
df = pd.read_csv('House_Rent_Dataset.csv')

# Convert 'Posted On' column to datetime
df['Posted On'] = pd.to_datetime(df['Posted On'])

# Group by 'Posted On' and count the number of houses rented
df_grouped = df.groupby('Posted On').size().reset_index(name='Count')

# Plot line graph
plt.figure(figsize=(12,6))
sns.lineplot(data=df_grouped, x='Posted On', y='Count')

plt.title('the amount of houses rented plotted against the date')
plt.xlabel('Date')
plt.ylabel('Number of Houses Rented')

plt.show()
```
