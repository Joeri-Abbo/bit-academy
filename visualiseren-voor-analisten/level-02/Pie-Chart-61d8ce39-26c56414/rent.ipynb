Helaas heeft seaborn (nog) geen pie chart functie, daarom gaan we matplotlib gebruiken. Het grote verschil tussen matplotlib en seaborn is dat seaborn gebouwd is om met pandas dataframes samen te werken. Matplotlib werkt ook met andere datastructuren als input, zoals een list.

Import dependencies die nodig zijn om de visualisatie te maken.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

Laad de data in

```python
data = pd.read_csv('House_Rent_Dataset.csv')
```

Maak de visualisatie

```python
huizen_per_stad = data['City'].value_counts()

steden = huizen_per_stad.index
aantal_huizen = huizen_per_stad.values

plt.pie(aantal_huizen, labels=steden, autopct='%1.1f%%')
plt.title('Verdeling van het aantal huizen per stad')
plt.axis('equal')
plt.show()
```
