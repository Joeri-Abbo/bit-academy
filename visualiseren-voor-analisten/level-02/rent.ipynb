### Importeren van csv

```python
house_df = pd.read_csv("House_Rent_Dataset.csv")
house_df.head(100)
```

### Importeren van seaborn

```python
import seaborn as sns
sns.__version__
house_df.shape
```

### Analyseren van de data en plotten van een historgram met seaborn

```python
#sns.set_style("darkgrid")
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.xlabel('Rent(INR)')
plt.ylabel("Number of apartments")
plt.title("Histogram of the Monthly rent of apartments")
sns.histplot(house_df.Rent)
```

### Tweede histogram met meer bins

```python
plt.figure(figsize=(10, 10))
plt.xlabel('Rent(INR)')
plt.ylabel("Number of apartments")
plt.title("Histogram of the Monthly rent of apartments")
sns.histplot(house_df.Rent, bins=50)
```

### De voordelen van een hoger aantal bins:
- Je kan veel meer detail zien. Je kan bepaalde pieken en dalen zien. Die je zou kunnen missen
- Je kunt meer informatie halen uit de histogram. Bijvoorbeeld groepen of outliers.

### De nadelen:
- Er is geen overzicht meer, dus teveel data
- Meer nadruk op outliers
- Ruis en fluctuaties.

# Violinplot

Gebruiken van violinplots

## 1. Maak een nieuw dataframe met alleen huizen in staat 1 en 2

```python
df_12 = house_df[(house_df['State'] == 1) | (house_df['State'] == 2)]
```

## 2. Maak van het nieuwe DataFrame-object een violinplot, met Rent op de y-as en City op de x-as;

```python
plt.figure(figsize=(10, 10))
sns.violinplot(x=df_12['City'], y=df_12['Rent'])
plt.xlabel('City')
plt.ylabel('Rent')
plt.title('Violinplot of Rent by City')

plt.show()
```

## 3. Zorg dat in de violinplot de huizen van verschillende staten naast elkaar geplaatst worden.

```python
plt.figure(figsize=(10, 10))
sns.violinplot(x=df_12['City'], y=df_12['Rent'], hue=df_12['State'], split=True, legend=False)
plt.xlabel('City')
plt.ylabel('Monthly Rent (INR)')
plt.title("The yearly rent of apartments in India plotted against the city in which they are located, grouped by the house's state")

# Create a dictionary to map the original labels to desired labels
state_labels = {1: 'Average State', 2: 'Worn Down State'}

# Create custom legend handles and labels
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8) for color in sns.color_palette()]
legend_labels = [state_labels[label] for label in df_12['State'].unique()]

# Display the legend
plt.legend(legend_handles, legend_labels, title='State')

plt.show()
```

## 4. Leg uit of de stelling klopt of niet.

De violinplot geeft mooi weer hoe de verdeling is per stad. Als we beginnen bij Kokata zien we dat worn down state duidelijk een lagere huurprijs heeft. De meeste huren liggen rond 8.000 INR en lopen ongeveer op tot een max 18.000 INR. Die van average state huizen lopen op tot over de 50.000 INR. Dit komt ongeveer overeen met alle andere steden behalve Hyderabad. Daar zie je dat bij beide "states" de meeste huurprijzen rond 12 a 13 duizen liggen.

## 5. Maak dezelfde plot van stap 3, maar schaal nu de oppervlaktes op basis van het aantal meetpunten.

```python
plt.figure(figsize=(10, 10))
sns.violinplot(x=df_12['City'], y=df_12['Rent'], hue=df_12['State'], split=True, legend=False, scale ='count')
plt.xlabel('City')
plt.ylabel('Monthly Rent (INR)')
plt.title("The yearly rent of apartments in India plotted against the city in which they are located, grouped by the house's state")

# Create a dictionary to map the original labels to desired labels
state_labels = {1: 'Average State', 2: 'Worn Down State'}

# Create custom legend handles and labels
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8) for color in sns.color_palette()]
legend_labels = [state_labels[label] for label in df_12['State'].unique()]

# Display the legend
plt.legend(legend_handles, legend_labels, title='State')

plt.show()
```

## 6. Voeg een swarmplot toe aan de laatste violinplot, waarop de verdeling van de datapunten te zien is. Gebruik palette=['black'] om de punten zichtbaar te houden.

```python
plt.figure(figsize=(10, 10))
sns.violinplot(x=df_12['City'], y=df_12['Rent'], hue=df_12['State'], split=True, legend=False, scale ='count')
sns.swarmplot(x=df_12['City'], y=df_12['Rent'], color='black', size=1, alpha=0.7, dodge=True, hue=df_12['State'])
plt.xlabel('City')
plt.ylabel('Monthly Rent (INR)')
plt.title("The yearly rent of apartments in India plotted against the city in which they are located, grouped by the house's state")

# Create a dictionary to map the original labels to desired labels
state_labels = {1: 'Average State', 2: 'Worn Down State'}

# Create custom legend handles and labels
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8) for color in sns.color_palette()]
legend_labels = [state_labels[label] for label in df_12['State'].unique()]

# Display the legend
plt.legend(legend_handles, legend_labels, title='State')

plt.show()
```

## 7. Probeer op basis van de laatste grafiek te verklaren waarom Hyderabad niet dezelfde trend vertoont als de andere steden.

Bij de andere steden is het aanbod van huizen met verschillende states niet evenredig verdeeld. Je ziet een veel groter aanbod van "Worn Down" huizen. Mumbai is een uitzondering waar het aantal average state huizen ver in de meerderheid is en er een laag aanbod is van "worn down" huizen. Bij Hyderabad is dit in balans.

# Lineplot

In deze opdracht leer je hoe je een lijngrafiek maakt met seaborn.

## 1. Groupeer de data op basis van datum('Posted On')

```python
# Convert the 'Posted On' column to datetime data type (if needed)
house_df['Posted On'] = pd.to_datetime(house_df['Posted On'])

from matplotlib import dates as mdates

# Group the data by date and count the number of rows per date
grouped_df = house_df.groupby(house_df['Posted On']).size().reset_index(name='Count')
```

## 2. Plot een lijngrafiek waarop het aantal verhuurde huizen per datum te zien is.

```python
from matplotlib import dates as mdates

# Group the data by date and count the number of rows per date
grouped_df = house_df.groupby(house_df['Posted On']).size().reset_index(name='Count')

plt.figure(figsize=(10, 10))
sns.lineplot(data=grouped_df, x='Posted On', y='Count')
plt.xlabel('Datum (YYYY-mm-dd)')
plt.ylabel('Number houses rented')
plt.title('The amound of houses rented plotted against the date')



# Adjust the x-axis tick frequency
ax = plt.gca()
locator = mdates.AutoDateLocator(minticks=7, maxticks=7)  # Set desired number of ticks
ax.xaxis.set_major_locator(mdates.DayLocator(interval=10))  # Set tick frequency to every 10 days
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Set the start date for the x-axis
start_date = pd.to_datetime('2022-04-14')
start_date_num = mdates.date2num(start_date)
ax.set_xlim(left=start_date_num)

plt.xticks(rotation=45)
plt.show()
```

# Pie Chart

Maken van een Pie chart in matplotlib

## Maak een taartdiagram waarin je de verdeling van het aantal huizen per stad laat zien.

```python
# Count the number of houses per city
city_counts = house_df['City'].value_counts()

# Get the city names and corresponding counts
city_names = city_counts.index.tolist()
city_values = city_counts.values.tolist()

# Create the pie chart
plt.figure(figsize=(10, 10))
plt.pie(city_values, labels=city_names, autopct='%1.1f%%')
plt.title('Distribution of Houses across cities')

plt.axis('equal')  # Ensure the pie is drawn as a circle
plt.show()
```

## 2. Maak met matplotlib drie barplots; In elke plot staat het aantal huizen per stad van Ã©Ã©n van de staten waarin een huis kan verkeren. Om deze plots in de goeie subplots te krijgen gebruik je in de barchart-functie het argument ax=.

```python
import numpy as np
# Create a color palette with enough colors for each city
color_palette = plt.cm.get_cmap('tab20', len(house_df['City'].unique()))

# Filter de data voor elke staat
state_1_df = house_df[house_df['State'] == 1]
state_2_df = house_df[house_df['State'] == 2]
state_3_df = house_df[house_df['State'] == 3]

# Bepaal de aantal huizen per stad voor elke staat
state_1_city_counts = state_1_df['City'].value_counts()
state_2_city_counts = state_2_df['City'].value_counts()
state_3_city_counts = state_3_df['City'].value_counts()

# Maak de figuur en assen aan
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

# Plot de barplots voor elke staat
state_1_ax = axes[0].bar(state_1_city_counts.index, state_1_city_counts.values, color=color_palette(np.arange(len(state_1_city_counts))))
axes[0].set_ylabel('The number of worn down houses in each city')
axes[0].set_xlabel('Rent (INR)')
axes[0].set_title('Staat 1')

state_2_ax = axes[1].bar(state_2_city_counts.index, state_2_city_counts.values, color=color_palette(np.arange(len(state_2_city_counts))))
axes[1].set_ylabel('Number of houses')
axes[1].set_xlabel('Rent (INR)')
axes[1].set_title('The number of averagly maintained houses in each city')

state_3_ax = axes[2].bar(state_3_city_counts.index, state_3_city_counts.values, color=color_palette(np.arange(len(state_3_city_counts))))
axes[2].set_ylabel('Number of new or as new houses in each city')
axes[2].set_xlabel('Rent (INR)')
axes[2].set_title('As new')

# Aanpassen van de layout van de subplots
fig.tight_layout()

# Toon de figuur
plt.show()
```
