# Scatterplot
In deze opdracht gaan we aan de slag met Scatterplot.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Lees het CSV-bestand in

```python
data = pd.read_csv('House_Rent_Dataset.csv')
```

Scatterplot met Size op de x-as en Rent op de y-as

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Size'], data['Rent'])
plt.xlabel('Size')
plt.ylabel('Rent')
plt.title('The montly rent of apartments plotted against their size')
plt.show()
```

De stelling: Hoe groter het huis, hoe hoger de huurpijs is juist!


Scatterplot met Size op de x-as, Rent op de y-as en kleuren gebaseerd op State

```python
state_labels = {1: 'Worn Down', 2: 'Average State', 3: 'As New'}
plt.figure(figsize=(10, 6))
scatter = plt.scatter(data['Size'], data['Rent'], c=data['State'], cmap='viridis')
plt.xlabel('Size')
plt.ylabel('Rent')
plt.title('The montly rent of apartments plotted against their size and grouped by their state')
legend_labels = [state_labels[val] for val in set(data['State'])]
plt.legend(handles=scatter.legend_elements()[0], labels=legend_labels, loc='upper right')
plt.show()
```

De stelling: De state van het huis heeft daar niks mee te maken en is dus onjuist.
Je kan bij deze plot goed zien dat de nieuwe huizen vaak duurder en groter zijn.
De gene met wat lagere huren zijn vaak voor de oudere huizen.
