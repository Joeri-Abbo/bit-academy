```python
!{sys.executable} -m pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Maak een boxplot met State op de x-as en Rent op de y-as;

```python
data = pd.read_csv('House_Rent_Dataset.csv')
```

```python
plt.figure(figsize=(10, 10))
sns.boxplot(x=data.State, y=data.Rent)
plt.xlabel('State')
plt.ylabel('Rent (INR)')
plt.title('The monthly rent of apartments plotted against the house´s state')

plt.show()
```

2. Zoek uit of stelling 1 klopt of niet.
Stelling 1:
Alle huizen met een staat 3 hebben een hogere huurprijs dan 75% van de huizen in staat 1.

Antwoord:
Deze bewering is onjuist, duidelijk te zien aan de snorharen van staat 3 dat er aanzienlijke overlapping is met staat 1. Er wordt tevens geclaimd dat het betrekking heeft op alle woningen. Daarom klopt dit in dit geval niet.

3. Leg uit of stelling 2 klopt of niet. Gebruik de boxplot die je in stap 1 hebt gemaakt voor je uitleg.

Stelling 2:

Antwoord:
De mediaan van staat 3 bevindt zich rond de 40.000. Het 75e percentiel ligt rond de 28.000. Daarom is stelling 2 inderdaad juist.

## 4. Maak een boxplot zoals in stap 1, maar maak nu onderscheid tussen gemeubileerde en ongemeubileerde huizen.

```python
plt.figure(figsize=(10, 10))
ax = sns.boxplot(x=data.State, y=data.Rent, hue=data['Furnishing Status'])
ax.set_xticklabels(['Worn down', 'Average State', 'As new'])
plt.xlabel('Furnishing Status')
plt.ylabel('Rent (INR)')
plt.title('The monthly rent of apartments plotted against the house´s state and grouped by their furnishing status')

plt.show()
```

5. Vergelijk de twee boxplots (uit stap 1 en 4) met elkaar, en leg uit of gemeubileerde huizen over het algemeen een andere huurprijs hebben dan niet-gemeubileerde huizen.

Het is overduidelijk dat gemeubileerde woningen een hogere mediaan hebben. Hoewel de whiskers bij "versleten" en "als nieuw" iets langer zijn, is het algemene patroon dat gemeubileerde woningen inderdaad een hogere huurprijs hebben in vergelijking met niet-gemeubileerde woningen.
